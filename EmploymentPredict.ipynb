```python
min_max = MinMaxScaler(feature_range=(0,1))
columns_to_normalize = ['Experience','Salary']

df_ready[columns_to_normalize] = min_max.fit_transform(df_ready[columns_to_normalize])
df_ready.head(10)
```

# Modelling, and predicting salaries

```python
lr = linear_model.LinearRegression()
logistic_reg = linear_model.LogisticRegression()
```

```python
X = df_ready[['Experience', 'Education_Encoded', 'AI_risk_Encoded',
       'Status_Encoded', 'Location_Delhi', 'Location_Hyderabad',
       'Location_Mumbai', 'Location_Rural', 'Location_Suburban',
       'Location_Urban', 'Industry_Finance', 'Industry_Fintech',
       'Industry_Healthcare', 'Industry_Retail', 'Industry_Technology',
       'Ages_25-34', 'Ages_35-44', 'Ages_45-54', 'Ages_55+']] # Features
y = df_ready['Salary']                 # Target variable

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
linear_model = lr.fit(X_train,y_train)

y_pred = linear_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# 3. Calculate R-squared
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2) Score: {r2:.4f}")
```

```python
salary_scaler = MinMaxScaler() 

salary_scaler.fit(df_ready[['Salary']])

real_y_test = salary_scaler.inverse_transform(np.array(y_test).reshape(-1, 1))
real_y_pred = salary_scaler.inverse_transform(np.array(y_pred).reshape(-1, 1))


# ---------------------------------------------------------
# STEP 3: Create a clean DataFrame to see the actual differences!
# ---------------------------------------------------------
# .flatten() converts the 2D array back into a flat list for the table
results_df = pd.DataFrame({
    'Actual Salary ($)': real_y_test.flatten(),
    'Predicted Salary ($)': real_y_pred.flatten()
})

# Calculate the real-world dollar error for every single guess!
results_df['Difference ($)'] = results_df['Predicted Salary ($)'] - results_df['Actual Salary ($)']

# Round the numbers so they look like real money
results_df = results_df.round(2)

print("--- REAL WORLD DOLLAR DIFFERENCES ---")
print(results_df.head(10))
```
